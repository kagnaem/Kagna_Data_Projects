# Running correlation for only numeric variables



```python
#Pearson correlation

df.corr(numeric_only=True)
```

Running Spearman Correlation wit Gradient background

```python

(df
 .corr(method='spearman', numeric_only=True)
 .style
 .background_gradient()
)
```

```python
#spearman correlation with style
(df
 .corr(method='spearman', numeric_only=True)
 .style
 .background_gradient(cmap='RdBu')
)
```
